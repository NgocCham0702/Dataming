```python
import sqlite3

conn = sqlite3.connect("itviec_full.db")
cur = conn.cursor()

# 1) Lấy 5 job_id đang có is_detailed_crawled = 2
cur.execute("""
    SELECT job_id
    FROM job_postings
    WHERE is_detailed_crawled = 2
    ORDER BY job_id
    LIMIT 5;
""")
rows = cur.fetchall()
ids = [r[0] for r in rows]

if ids:
    # 2) Update 5 dòng đó về 0
    sql = f"""
        UPDATE job_postings
        SET is_detailed_crawled = 0
        WHERE job_id IN ({",".join("?" for _ in ids)});
    """
    cur.execute(sql, ids)
    conn.commit()

conn.close()
```
